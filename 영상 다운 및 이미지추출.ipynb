```python
#   
import os
import cv2

#   
video_path = '/Users/cyr95/Desktop// (2019).mp4'

#   
save_directory = '/Users/cyr95/anaconda3/envs/VT/youtube_image2/ama_image3'

#         (: 10  1)
interval = 1  #  

#    
video = cv2.VideoCapture(video_path)

#   
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

frame_number = 0
time_elapsed = 0.0
frame_rate = video.get(cv2.CAP_PROP_FPS)

while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break

    #   
    time_elapsed = frame_number / frame_rate

    #       
    if (interval > 1 and frame_number % interval == 0) or (interval <= 1 and time_elapsed >= interval):
        image_path = os.path.join(save_directory, f'frame_{frame_number}.jpg')
        cv2.imwrite(image_path, frame)
        print(f'Saved image: {image_path}')

    frame_number += 1

#  
video.release()
cv2.destroyAllWindows()
```

```python
from googleapiclient.discovery import build
from pytube import YouTube

# YouTube API 
api_key = "AIzaSyD9V4ctSuYjy7BwPljhxgm4oEeM3ZeSKhs"

# YouTube 
search_query = "golf+pro+swing"

#    (: )
max_duration = 600  # 10

#   
max_downloads = 50

# YouTube Data API  
youtube = build('youtube', 'v3', developerKey=api_key)

# YouTube  
search_response = youtube.search().list(
    q=search_query,
    part='id',
    maxResults=max_downloads  #    
).execute()

#    ID 
video_ids = []
for item in search_response['items']:
    if item['id']['kind'] == 'youtube#video':
        video_ids.append(item['id']['videoId'])

#   
for video_id in video_ids:
    try:
        # YouTube  URL 
        video_url = f"https://www.youtube.com/watch?v={video_id}"

        # YouTube  
        yt = YouTube(video_url)

        #   
        duration = int(yt.length)

        #       
        if duration <= max_duration:
            #     
            streams = yt.streams.all()

            #   
            for stream in streams:
                #  
                stream.download(output_path="/Users/cyr95/Desktop/ ")
            print(" :", yt.title)
        else:
            print("  -    :", yt.title)
    except Exception as e:
        print(" :", e)
```
